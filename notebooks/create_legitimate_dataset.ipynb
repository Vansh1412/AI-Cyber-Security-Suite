```python
import pandas as pd

# Read downloaded Tranco CSV
df = pd.read_csv(
    "top-1m.csv",
    names=["rank", "domain"]
)

# Keep only domain column
df = df[["domain"]]

# Convert to URL
df["url"] = "https://" + df["domain"]

# Legitimate label
df["label"] = 0

# Remove domain column
df = df[["url", "label"]]

# Save
df.to_csv(
    "legitimate_urls.csv",
    index=False
)

print(df.head())
```
